```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v=np.linspace(0,1,60)
w=np.linspace(0,2,80)
f=np.sin(v)
g=np.exp(-w**2)
plt.plot(v,f)
plt.plot(w,g)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot of Functions')
plt.legend(['sin(x)', 'exp(-x^2)'])
plt.grid()
plt.show()
```

```python
plt.plot(v,f,'r--',w,g,'x')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot of Functions')
plt.legend(['sin(x)', 'exp(-x^2)'])
plt.grid()
plt.xlim(0,1)
plt.show()
```

```python
def add(a,b):
    return a+b
def trapezoidal_rule(f,a,b,N):
    x=np.linspace(a,b,N+1)
    y=f(x)
    h=(b-a)/N
    integral=(h/2)*(y[0]+2*np.sum(y[1:N])+y[N])
    return integral
```
